```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar los datos
df = pd.read_csv('Spotify.csv', sep=';', decimal=',')

# 1. Distribución de géneros musicales
plt.figure(figsize=(10, 8))
genre_counts = df['Genero'].value_counts()
plt.pie(genre_counts, labels=genre_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribución de Géneros Musicales en Spotify')
plt.show()

# 2. Distribución de tipos de plataforma (Free vs Premium)
plt.figure(figsize=(8, 8))
platform_counts = df['Tipo Plataforma'].value_counts()
plt.pie(platform_counts, labels=platform_counts.index, autopct='%1.1f%%', colors=['#1DB954', '#191414'])
plt.title('Distribución de Tipos de Suscripción en Spotify')
plt.show()

# 3. Distribución de oyentes por país (top 10)
plt.figure(figsize=(10, 8))
country_listeners = df.groupby('Pais')['Oyentes Mensuales Millones'].sum().sort_values(ascending=False).head(10)
plt.pie(country_listeners, labels=country_listeners.index, autopct='%1.1f%%')
plt.title('Distribución de Oyentes Mensuales por País (Top 10)')
plt.show()
```
